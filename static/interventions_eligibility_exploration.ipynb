```python
def query_handler(query_execution):
    d, a = {}, []
    for rowproxy in query_execution:
        # rowproxy.items() returns an array like [(key0, value0), (key1, value1)] > thank you stackoverflow!
        for column, value in rowproxy.items():
            # build up the dictionary
            d = {**d, **{column: value}}
        a.append(d)
    a_df = pd.DataFrame(a)
    connection.close()
    return a_df
```

```python
query = engine.execute(f"Select * From Inclusion_Criteria")
```

```python
inclusion_df = query_handler(query)
inclusion_df
```

```python
query = engine.execute(f"Select * From Exclusion_Criteria")
exclusion_df = query_handler(query)
exclusion_df
```
